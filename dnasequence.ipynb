```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Specify the file path
file_path = r"C:\Users\jeffr\Downloads\TargetALS-WGS-all-IDs-2022Feb21.xlsx - Sheet1 (1).csv"

# Load the dataset into a Pandas DataFrame
dfWGS = pd.read_csv(file_path)

```

```python
print(dfWGS.head())
print("Number of rows in dfWGS:", dfWGS.shape[0])
```

```python
# Iterate over unique ExternalSubjectId values
print("Every duplicated value in WGS database and their CompBioIDs")
for external_subject_id in dfWGS['ExternalSubjectId'].unique():
    # Get all rows with the current ExternalSubjectId
    rows_with_subject_id = dfWGS[dfWGS['ExternalSubjectId'] == external_subject_id]
    
    # Get unique CompBioId values associated with this ExternalSubjectId
    unique_comp_bio_ids = rows_with_subject_id['CompBioID'].unique()
    
    # If there are multiple unique CompBioIds, print them
    if len(unique_comp_bio_ids) > 1:
        print("ExternalSubjectId:", external_subject_id)
        print("Different CompBioIds:")
        for comp_bio_id in unique_comp_bio_ids:
            print(comp_bio_id)
        print()
```

```python
file_path1 = r"C:\Users\jeffr\Downloads\TargetALS-RNASeq-all-IDs-2022Feb21.xlsx - Sheet1.csv"

# Load the dataset into a Pandas DataFrame
dfRNA = pd.read_csv(file_path1)
```

```python
# Iterate over unique ExternalSubjectId values
print("Every duplicated value in RNA database and their CompBioIDs")
for external_subject_id in dfRNA['ExternalSubjectId'].unique():
    # Get all rows with the current ExternalSubjectId
    rows_with_subject_id = dfRNA[dfRNA['ExternalSubjectId'] == external_subject_id]
    
    # Get unique CompBioId values associated with this ExternalSubjectId
    unique_comp_bio_ids = rows_with_subject_id['CompBioID'].unique()
    
    # If there are multiple unique CompBioIds, print them
    if len(unique_comp_bio_ids) > 1:
        print("ExternalSubjectId:", external_subject_id)
        print("Different CompBioIds:")
        for comp_bio_id in unique_comp_bio_ids:
            print(comp_bio_id)
        print()
```

```python
import numpy as np

# Identify duplicated ExternalSubjectId values
duplicated_subject_ids = dfRNA['ExternalSubjectId'][dfRNA['ExternalSubjectId'].duplicated()].unique()

# Convert to array
duplicated_subject_ids_array = np.array(duplicated_subject_ids)

# Print the array
print("Array with duplicated ExternalSubjectId values:", duplicated_subject_ids_array)
len(duplicated_subject_ids_array)
```

```python
# Extract unique ExternalSubjectId values from both DataFrames
wgs_subject_ids = set(dfWGS['ExternalSubjectId'])
rna_subject_ids = set(dfRNA['ExternalSubjectId'])

# Find common values (intersection) between the two sets
common_subject_ids = wgs_subject_ids.intersection(rna_subject_ids)

# Convert the set to a list if needed
common_subject_ids_list = list(common_subject_ids)

# Print the common values
print("Common ExternalSubjectId values between dfWGS and dfRNA:")
print(common_subject_ids_list)
```

```python
# Merge dfWGS and dfRNA on 'ExternalSubjectId'
merged_df = pd.merge(dfWGS, dfRNA, on='ExternalSubjectId', suffixes=('_WGS', '_RNA'))

# Filter rows where 'CompBioID' values are different
different_comp_bio_ids = merged_df[merged_df['CompBioID_WGS'] != merged_df['CompBioID_RNA']]

# Extract unique 'ExternalSubjectId' values from the filtered DataFrame
different_subject_ids = different_comp_bio_ids['ExternalSubjectId'].unique()

# Print the values
print("ExternalSubjectId values in both dfWGS and dfRNA with different CompBioId in either of the datasets:")
print(different_subject_ids)
```
